```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.cluster import *
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_log_error
```

```python
#import data
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
```

```python
def check_null(data):
    variables = data.columns.to_list()
    check_null = []
    for x in variables:
        nnull = data[x].isnull().sum()
        check_null.append([x,nnull])
    return check_null
#check_null(train)
check_null(test)
```

```python
train.shape
```

```python
#data_cleaning
def one_hot_encode(data):
    dum = pd.get_dummies(data.loc[:,['Sex','Embarked']])
    return data.join(dum).drop(['Sex','Embarked'], axis = 1)
def data_cleaning(data):
    data1 = data.drop(["Cabin","Ticket","Name","PassengerId"],axis=1)
    return one_hot_encode(data1)
train1 = data_cleaning(train).dropna()
test1 = data_cleaning(test).ffill()
train1
```

```python
x = train1.drop("Survived",axis=1)
y = train1['Survived']
x_train, x_test, y_train, y_test =train_test_split(x,y, test_size=0.25)
linear_model = RandomForestRegressor(n_estimators = 10,max_depth = 7, random_state = 1)
linear_model.fit(x_train, y_train)
y_fitted = linear_model.predict(x_train)
y_predicted = linear_model.predict(x_test)
train_rmsle_cases=np.sqrt(mean_squared_log_error(y_train,y_fitted))
test_rmsle_cases=np.sqrt(mean_squared_log_error(y_test,y_predicted))
train_rmsle_cases, test_rmsle_cases
```

```python
predicted = linear_model.predict(test1)
result = pd.DataFrame({"PassengerId":test["PassengerId"],
                      "Survived":predicted}).set_index("PassengerId")
result.to_csv("prediction.csv")
result
```
